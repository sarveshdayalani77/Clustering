```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
crime=pd.read_csv('/content/crime_data.csv')
```

```python
crime.head
```

```python
#Normalized data fuction
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

The norm_func function takes a dataset i as input and returns the dataset with its values normalized using the min-max normalization technique, ensuring they are scaled within the range [0, 1].


```python
df_norm=norm_func(crime.iloc[:,1:])
```

```python
df_norm
```

**Hierarichal Clustering**

```python
dendogram=sch.dendrogram(sch.linkage(df_norm,method='average'))
```

```python
#creating clusters
hc=AgglomerativeClustering(n_clusters=3,metric='euclidean',linkage='complete')
```

```python
hc
```

```python
y_hc=hc.fit_predict(df_norm)
```

This is used to fit the clustering algorithm to the normalized data and obtain the cluster labels

```python
y_hc
```

```python
crime['Cluster_id']=hc.labels_
```

**Kmeans**

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist

crime1=pd.read_csv('/content/crime_data.csv')
```

```python
#Normalized data fuction
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
df_norm=norm_func(crime1.iloc[:,1:])
```

```python
df_norm
```

```python
wcss=[]
for i in range (1,11):
  kmeans=KMeans(n_clusters=i)
  kmeans.fit(df_norm)
  wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title('elbow curve')
plt.xlabel('no of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
# selecting 4 clusters from above plot
model=KMeans(n_clusters=4)
model.fit(df_norm)
model.labels_
```

```python
x=pd.Series(model.labels_)
x
```

```python
crime1['clust']=x
crime1
```

```python
crime1.iloc[:,1:5].groupby(crime1.clust).mean()
```

**DBSCAN**

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
crime2=pd.read_csv('/content/crime_data.csv')
```

```python
crime2.head
```

```python
crime2.info()
```

```python
df=crime2.iloc[:,1:5]
```

```python
df
```

```python
df.values
```

```python
stscaler=StandardScaler().fit(df.values)
stscaler
```

```python
x=stscaler.transform(df.values)
x
```

```python
dbscan=DBSCAN(eps=2,min_samples=5)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
c1=pd.DataFrame(dbscan.labels_,columns=['Cluster'])
```

```python
c1
```
